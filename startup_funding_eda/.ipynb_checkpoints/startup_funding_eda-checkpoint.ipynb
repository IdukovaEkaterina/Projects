## Введение <a id='section_0'></a>

### Цели и задачи проекта

**Цель:** Провести исследовательский анализ данных о финансировании стартапов, понять закономерности и оценить перспективы выхода на рынок с покупкой и развитием компаний. 

**Задачи:**
1. Загрузить данные и познакомиться с их содержимым.
2. Провести предобработку данных.
3. Провести исследовательский анализ данных:
	- выделить группы компаний по срокам финансирования и сравнить их по количеству и объёму инвестиций;
	- классифицировать сегменты рынка на массовые, средние и нишевые;
	- определить типичные и аномальные значения объёмов финансирования;
	- сравнить популярность и объёмы разных типов финансирования;
	- проанализировать динамику раундов и объёмов инвестиций по годам, а также изменения в массовых сегментах рынка;
    - рассчитать долю возврата средств для разных типов финансирования и оценить её устойчивость.
4. Подготовить итоговые выводы и рекомендации.

### Описание данных

Исследование основано на двух наборах данных. Основной датасет содержит информацию о компаниях, рыночных сегментах, статусе, сроках и объёмах финансирования. Дополнительный датасет содержит данные о возвратах средств по типам финансирования за 2000–2014 годы.

Единицей наблюдения в основном датасете является компания. Основной показатель анализа — общий объём привлечённого финансирования `funding_total_usd`.

Датасеты получены из базы данных стартапов.

Название основного датасета — `cb_investments.zip`. Внутри архива один файл — `cb_investments.csv`.

Описание данных:
* `name` — название компании.
* `homepage_url` — ссылка на сайт компании.
* `category_list` — категории, в которых работает компания. Указываются через `|`.
* `market` — основной рынок или отрасль компании.
* `funding_total_usd` — общий объём привлечённых инвестиций в долларах США.
* `status` — текущий статус компании, например `operating`, `closed` и так далее.
* `country_code` — код страны, например USA.
* `state_code` — код штата или региона, например, CA.
* `region` — регион, например, SF Bay Area.
* `city` — город, в котором расположена компания.
* `funding_rounds` — общее число раундов финансирования.
* `participants` — число участников в раундах финансирования.
* `founded_at` — дата основания компании.
* `founded_month` — месяц основания в формате `YYYY-MM`.
* `founded_quarter` — квартал основания в формате `YYYY-QN`.
* `founded_year` — год основания.
* `first_funding_at` — дата первого финансирования.
* `mid_funding_at` — дата среднего по времени раунда финансирования.
* `last_funding_at` — дата последнего финансирования.
* `seed` — сумма инвестиций на посевной стадии.
* `venture` — сумма венчурных инвестиций.
* `equity_crowdfunding` — сумма, привлечённая через долевой краудфандинг.
* `undisclosed` — сумма финансирования нераскрытого типа.
* `convertible_note` — сумма инвестиций через конвертируемые займы.
* `debt_financing` — сумма долгового финансирования.
* `angel` — сумма инвестиций от бизнес-ангелов.
* `grant` — сумма полученных грантов.
* `private_equity` — сумма инвестиций в виде прямых (частных) вложений.
* `post_ipo_equity` — сумма финансирования после IPO.
* `post_ipo_debt` — сумма долгового финансирования после IPO.
* `secondary_market` — сумма сделок на вторичном рынке.
* `product_crowdfunding` — сумма, привлечённая через продуктовый краудфандинг.
* `round_A` — `round_H` — сумма инвестиций в соответствующем раунде.

Название дополнительного датасета — `cb_returns.csv`. Он содержит суммы возвратов по типам финансирования в миллионах долларов по годам.

Описание данных:
* `year` — год возврата средств.
* `seed` — сумма возвратов от посевных инвестиций.
* `venture` — сумма возвратов от венчурных инвестиций.
* `equity_crowdfunding` — сумма, возвращённая по долевому краудфандингу.
* `undisclosed` — сумма возвратов нераскрытого типа.
* `convertible_note` — сумма возвратов через конвертируемые займы.
* `debt_financing` — сумма возвратов от долгового финансирования.
* `angel` — сумма возвратов бизнес-ангелам.
* `grant` — сумма возвратов по грантам.
* `private_equity` — сумма возвратов прямых (частных) вложений.
* `post_ipo_equity` — сумма возвратов от IPO.
* `post_ipo_debt` — сумма возвратов от долгового IPO.
* `secondary_market` — сумма возвратов от сделок на вторичном рынке.
* `product_crowdfunding` — сумма возвратов по продуктовому краудфандингу.

### 2.2. Типы данных <a id='section_2.2'></a>

На стадии знакомства с данными было установлено:
- Столбец `funding_total_usd` датасета `investments` хранится как текст, необходимо убрать выделение разрядов и привести столбец к числовому типу.
- В столбцах `founded_at`, `first_funding_at`, `mid_funding_at`, `last_funding_at` датасета `investments` даты представлены строками, их нужно преобразовать в формат `datetime`.
- В столбцах `founded_year`,`funding_rounds`, `participants` нужно обработать пропуски и привести данные к целочисленному типу (Будет осуществлено в разделе [2.4 Наличие пропусков в данных](#section_2.4)).
- Для дальнейшего анализа столбец `year` датасета `returns` нужно сделать индексом всего датасета.
- Для всех столбцов датасета `returns` тип данных можно оптимизировать.

```python
# Очищаем столбец funding_total_usd и приводим его к числовому типу
investments['funding_total_usd'] = investments['funding_total_usd'].str.replace(',', '')
investments['funding_total_usd'] = pd.to_numeric(investments['funding_total_usd'], errors='coerce')
```

```python
# Приводим столбцы с датами к типу данных datetime
for column in ['founded_at', 'first_funding_at', 'mid_funding_at', 'last_funding_at']:
    investments[column] = pd.to_datetime(investments[column], errors='coerce')
```

```python
# В датасете returns делаем столбец year индексом всего датасета
returns = returns.set_index('year')
returns.head()
```

```python
# Оптимизируем типы данных в датафрейме returns
for column in returns.columns:
    returns[column] = pd.to_numeric(returns[column], downcast='float')
```

```python
# Проверим типы данных в датафрейме investments
investments.dtypes
```

```python
# Проверим типы данных в датафрейме returns
returns.dtypes
```

### 2.3 Явные и неявные дубликаты в данных <a id='section_2.3'></a>

Проверим данные на наличие явных и неявных дубликатов. Для этого проведём нормализацию текстовых признаков, после чего выполним поиск повторяющихся записей и при необходимости устраним выявленные дубликаты.

```python
# Проверяем наличие явных дубликатов в данных до нормализации
investments.duplicated().sum()
```

```python
# Процент явных дубликатов в данных
(investments.duplicated().mean() * 100).round(2)
```

```python
# Проверяем наличие явных дубликатов в данных
returns.duplicated().sum()
```

```python
# Удаляем явные дубликаты в данных
investments = investments.drop_duplicates()
```

```python
# Проводим нормализацию текстовых признаков 
for column in ['name', 'homepage_url', 'market', 'category_list', 'region', 'city']:
    investments[column] = investments[column].str.lower().str.strip()
```

```python
# Проверяем наличие явных дубликатов в данных после нормализации
investments.duplicated().sum()
```

```python
# Проверяем наличие неявных дубликатов в данных
investments.duplicated(subset=['name', 'homepage_url']).sum()
```

```python
# Процент неявных дубликатов в данных
(investments.duplicated(subset=['name', 'homepage_url']).mean() * 100).round(2)
```

```python
# Удаляем выявленые дубликаты в данных
investments = investments.drop_duplicates(subset=['name', 'homepage_url'])
```

```python
# Проверяем корректность удаления дубликатов
investments.duplicated(subset=['name', 'homepage_url']).sum()
```

```python
# Посчитаем количество удалённых строк в абсолютном и относительном значениях
final_row_count = len(investments)
initial_row_count= len(investments_copy)
deleted_rows_count = initial_row_count - final_row_count
deleted_percent = deleted_rows_count / initial_row_count * 100
display(f'Количество удаленных строк: {deleted_rows_count}')
display(f'Процент удалённых строк: {round(deleted_percent,2)} %')
```

Проверка показала, что в данных присутствовали 4855 явных и 14 неявных дубликата по сочетанию названия и ссылки на сайт компании. Вероятно, дубликаты возникли из-за особенностей сбора данных и дублированием карточек в источнике. Поскольку для дальнейшего анализа важно исключить повторный учёт одних и тех же компаний, такие дубликаты были удалены и оставлено по одной записи для каждой пары `name` + `homepage_url`.
Итого было удалено 4869 строк или 8.97 % данных.

Вероятнее всего, отсутствие данных связано с неполнотой источников или потерей данных при сборе. Доля таких записей мала и не оказывает статистически значимого влияния на выборку. При этом отсутствие ключевых идентифицирующих признаков делает данные строки малополезными для анализа и затрудняет их корректную интерпретацию. В связи с этим оптимальным решением является удаление данных строк, что не повлияет на качество и репрезентативность анализа.

```python
# Загружаем библиотеку для визуализации пропусков
!pip install -q missingno
import missingno as msno

# Визуализируем пропуски в данных
msno.matrix(investments)
plt.title('Матрица пропусков в датасете investments')
plt.show()
```

```python
# Посмотрим на взаимосвязь пропусков в разных столбцах
msno.heatmap(
    investments,
    labels=False
)

plt.title('Корреляция пропусков между столбцами')
plt.show()
```

Матрица пропусков показала, что отсутствующие значения распределены неравномерно. В данных есть как отдельные пропуски, так и крупные блоки совместно отсутствующих значений. Особенно заметен блок пропусков в признаках даты основания компании: `founded_at`, `founded_month`, `founded_quarter` и `founded_year`. Это указывает на системный характер пропусков, для части компаний информация о дате основания отсутствует полностью, а не пропущена случайным образом в отдельных строках. Подобная структура может свидетельствовать о различиях между источниками данных или неполном заполнении отдельных групп компаний в исходной базе.

Тепловая карта корреляции пропусков показала, что часть отсутствующих значений  связана между собой. Наиболее выраженная связь наблюдается между признаками рыночной категории (`category_list`, `market`), географии компании (`country_code`, `state_code`, `region`, `city`) и даты основания (`founded_at`, `founded_month`, `founded_quarter`, `founded_year`).

Это означает, что для части компаний в исходной базе отсутствуют не отдельные значения, а целые блоки информации. Например, при отсутствии данных о местоположении компании одновременно часто отсутствуют сведения о стране, регионе и городе, а при отсутствии даты основания, производные признаки месяца, квартала и года основания. Можно предположить что наша выборка - это результат объединения двух таблиц.

Таким образом, пропуски в данных имеют преимущественно групповой и системный характер, что может быть связано с объединением данных из разных источников или неполным заполнением отдельных карточек компаний.

```python
# Заполняем mid_funding_at серединой интервала между первым и последним финансированием
mid_missing_mask = (
    investments['mid_funding_at'].isna()
    & investments['first_funding_at'].notna()
    & investments['last_funding_at'].notna()
)

investments.loc[mid_missing_mask, 'mid_funding_at'] = (
    investments.loc[mid_missing_mask, 'first_funding_at']
    + (
        investments.loc[mid_missing_mask, 'last_funding_at']
        - investments.loc[mid_missing_mask, 'first_funding_at']
    ) / 2
)
```

```python
# Посмотрим на оставшиеся пропуски в столбце mid_funding_at
investments.loc[investments['mid_funding_at'].isna(), ['first_funding_at', 'mid_funding_at', 'last_funding_at' ]]
```

```python
# Заполняем пропуски в столбце founded_year заглушкой '-1' и преобразовываем к типу int16
investments['founded_year'] = investments['founded_year'].fillna(-1).astype('int16')
```

```python
# Удаляем пропуски в столбце funding_total_usd
investments = investments.dropna(subset=['funding_total_usd'])
```

#### 3.1.1. Группы по срокам финансирования

Разделим компании на три группы:

* `Единичное финансирование` - был всего один раунд финансирования.
* `Срок финансирования до года` - между первым и последним раундом финансирования прошло не более года.
* `Срок финансирования более года` - между первым и последним раундом прошло больше года.

```python
# Рассчитаем длительность между первым и последним финансированием
investments['funding_period_days'] = (investments['last_funding_at'] - investments['first_funding_at']).dt.days
```

```python
# Выделяем группы в столбец funding_period_group
# Единичное финансирование
investments.loc[(investments['funding_rounds'] == 1), 'funding_period_group'] = 'Единичное финансирование'
# Срок до года
investments.loc[(investments['funding_rounds'] != 1) & 
    (investments['funding_period_days'] <= 365), 'funding_period_group'] = 'Срок финансирования до года'
# Срок более года
investments.loc[(investments['funding_rounds'] != 1) & 
    (investments['funding_period_days'] > 365), 'funding_period_group'] = 'Срок финансирования более года'
# Отдельно обрабатываем пропуски
investments.loc[investments['funding_period_days'].isna(),'funding_period_group'] = 'Срок финансирования неизвестен'
```

```python
# Рассчитаем количество и долю компаний по группам
company_period_table = pd.DataFrame({
    'Количество компаний': investments['funding_period_group'].value_counts(),
    'Доля группы, %': investments['funding_period_group'].value_counts(normalize=True)*100
})

company_period_table = company_period_table.sort_values('Количество компаний', ascending=False)
company_period_table
```

```python
# Рассчитаем объём и долю инвестиций по группам
funding_period_table = pd.DataFrame({
    'Объём привлечённых средств, USD': (
        investments.groupby('funding_period_group')['funding_total_usd'].sum()),
    'Доля группы, %': (
        investments.groupby('funding_period_group')['funding_total_usd'].sum() /
        investments['funding_total_usd'].sum()* 100)    
})

funding_period_table = funding_period_table.sort_values('Объём привлечённых средств, USD', ascending=False)
funding_period_table
```

```python
# Подготавливаем данные
company_data = company_period_table.drop('Срок финансирования неизвестен').copy()
funding_data = funding_period_table.drop('Срок финансирования неизвестен').copy()

# Переносим длинные подписи
company_data.index = [
    'Единичное\nфинансирование',
    'Срок финансирования\nболее года',
    'Срок финансирования\nдо года'
]

funding_data.index = [
    'Срок финансирования\nболее года',
    'Единичное\nфинансирование',
    'Срок финансирования\nдо года'
]

# Единая палитра цветов
funding_palette = {
    'Единичное\nфинансирование': '#6b9ac4',
    'Срок финансирования\nболее года': '#fc8d62',
    'Срок финансирования\nдо года': '#c06c84'
}

# Создаём контейнер для графиков и задаём его размер
fig, axes = plt.subplots(1, 2, figsize=(12, 4))

# Строим графики
for ax, data, title in zip(
    axes,
    [company_data, funding_data],
    [
        'Распределение компаний\nпо сроку финансирования',
        'Доля привлечённых средств\nпо срокам финансирования'
    ]
):
    ax.pie(
        data['Доля группы, %'],
        labels=data.index,
        autopct='%.1f%%',
        startangle=90,
        counterclock=False,
        colors=[funding_palette[label] for label in data.index],
        wedgeprops={'width': 0.50},
        labeldistance=1.2,
        pctdistance=0.75,
    )
    
    ax.set_title(title)

plt.tight_layout()
plt.show()
```

#### 3.1.2 Выделение средних и нишевых сегментов рынка

- более 120 компаний — `массовый сегмент`;
- от 35 до 120 компаний включительно — `средний сегмент`;
- менее 35 компаний — `нишевый сегмент`.

После классификации оставим названия только для массовых сегментов. Средние и нишевые сегменты объединим в группы `mid` и `niche`.

Массовые сегменты подходят для поиска более зрелых компаний с подтверждённым спросом, развитой инфраструктурой рынка и более высокой вероятностью масштабирования. Нишевые сегменты, несмотря на небольшое число компаний, могут быть перспективны для поиска недооценённых активов с потенциалом быстрого роста, однако требуют более глубокой проверки устойчивости бизнес-модели и инвестиционной привлекательности.


В целом структура рынка стартапов выглядит естественной для быстроразвивающейся инновационной среды. Большое количество нишевых сегментов показывает, что рынок активно экспериментирует с новыми направлениями, продуктами и бизнес-моделями. Такие сегменты создают пространство для появления новых идей, но большинство из них пока остаётся небольшими по числу компаний.

При этом основная масса компаний сосредоточена в ограниченном числе крупных сегментов. Это говорит о том, что часть направлений уже прошла этап первичного рыночного отбора, в них сформировался спрос, появились конкуренты, инвесторы и инфраструктура для дальнейшего роста.

Такое распределение поддерживает развитие стартап-рынка, нишевые сегменты обеспечивают приток новых идей и экспериментов, а массовые сегменты концентрируют компании, капитал и конкуренцию. В результате рынок одновременно сохраняет инновационность и формирует более зрелые направления для масштабирования бизнеса.

### 3.2. Работа с выбросами и анализ <a id='section_3.2'></a>

#### 3.2.1. Анализируем и помечаем выбросы в каждом из сегментов

- Оценим распределение общего объёма финансирования и определим какой размер общего финансирования для одной компании будет типичным, а какой - выбивающимся.<br>
- Определим компании с аномальным объёмом общего финансирования с помощью метода IQR отдельно по каждому сегменту.<br>
- Определим сегменты рынка с наибольшей долей компаний, получивших аномальное финансирование.

```python
# Изучим статистические показатели столбца funding_total_usd
investments['funding_total_usd'].describe()
```

```python
# Строим гистограмму
ax = sns.histplot(
    (investments.loc[investments['funding_total_usd'] <= 25_000_000, 'funding_total_usd']) / 1_000_000,
    bins=50,
    alpha=1
)

# Добавляем заголовок и подписи
ax.set_title('Распределение общего объёма финансирования компаний (ограничение: 25 млн $)')
ax.set_xlabel('Общий объём финансирования, млн $')
ax.set_ylabel('Частота')

# Отображаем график
plt.show()
```

```python
# Строим диаграмму размаха без отображения выбросов
ax = sns.boxplot(
    x=(investments['funding_total_usd'] / 1_000_000),
    showfliers=False,
        width=0.3
)

# Добавляем заголовок и подпись оси
ax.set_title('Распределение общего объёма финансирования компаний без учёта выбросов')
ax.set_xlabel('Общий объём финансирования, млн $')

# Отображаем график
plt.show()
```

Общий объём финансирования распределён крайне неравномерно: среднее значение значительно выше медианы, что указывает на влияние крупных выбросов. Медианная компания привлекла около 2 млн долларов, тогда как среднее значение составляет около 15.9 млн долларов.

Гистограмма с ограничением до 25 млн долларов показывает, что основная масса компаний находится в зоне относительно небольших объёмов финансирования. Диаграмма размаха без отображения выбросов подтверждает правостороннюю асимметрию: типичный диапазон смещён к меньшим значениям, а крупные объёмы финансирования встречаются заметно реже.

В качестве типичного интервала можно принять диапазон между первым и третьим квартилем: от 350 тыс. до 10 млн долларов. Значения выше этого диапазона уже выглядят менее типичными, а формально аномальные значения будут определены далее методом IQR отдельно по каждому рыночному сегменту.

```python
# Рассчитываем квартильные значения по каждому сегменту
q1 = investments.groupby('market')['funding_total_usd'].transform(lambda x: x.quantile(0.25))
q3 = investments.groupby('market')['funding_total_usd'].transform(lambda x: x.quantile(0.75))

# Рассчитываем межквартильный размах
iqr = q3 - q1

# Рассчитываем границы выбросов
lower_bound = q1 - 1.5 * iqr
upper_bound = q3 + 1.5 * iqr

# Помечаем компании с аномальным объёмом финансирования
investments['funding_outlier'] = (
    (investments['funding_total_usd'] < lower_bound) |
    (investments['funding_total_usd'] > upper_bound)
)

# Проверяем количество обычных и аномальных значений
investments['funding_outlier'].value_counts()
```

```python
# Определим топ-10 сегментов с наибольшей долей компаний, получивших аномальное финансирование
top_outlier_segments = pd.DataFrame({
    'Кол-во компаний': investments.groupby('market')['funding_outlier'].count(),
    'Кол-во аномалий': investments.groupby('market')['funding_outlier'].sum(),
    'Доля аномалий, %': (investments.groupby('market')['funding_outlier'].mean() * 100)
})

top_outlier_segments = top_outlier_segments.sort_values('Доля аномалий, %',ascending=False)
top_outlier_segments.head(10)
```

```python
# Строим линейчатую диаграмму
ax = sns.barplot(
    x=top_outlier_segments.head(10)['Доля аномалий, %'],
    y=top_outlier_segments.head(10).index,
    color='#66c2a5'
)

# Добавляем заголовок и подписи осей
ax.set_title('Топ-10 сегментов по доле аномального финансирования')
ax.set_xlabel('Доля компаний с аномальным финансированием, %')
ax.set_ylabel('Рыночный сегмент')

# Добавляем подписи вручную
for index, value in enumerate(top_outlier_segments.head(10)['Доля аномалий, %']):
    ax.text(
        value + 0.3,
        index,
        f'{value:.1f}%',
        va='center'
    )

# Отображаем график
plt.xlim(0, 20)
plt.show()
```

```python
# Рассчитываем общую долю компаний с аномальным финансированием
outlier_share_total = investments['funding_outlier'].mean() * 100
print(f'Доля компаний с аномальным финансированием: {outlier_share_total:.2f}%')
```

Метод IQR позволил определить компании с аномальным объёмом финансирования отдельно внутри каждого рыночного сегмента. Такой подход корректнее общего расчёта по всей выборке, так как типичный размер финансирования может сильно отличаться между разными рынками.

Всего аномальное финансирование получили 5 240 компаний, или около 12.8% выборки. Это означает, что значительная часть компаний привлекла объём средств, заметно превышающий типичный уровень внутри своего сегмента.

Наибольшая доля аномального финансирования наблюдается в сегментах `real estate`, `entertainment`, `consulting`, `search` и `cloud computing`: в каждом из них доля таких компаний составляет около 16–17%. Это может указывать на наличие отдельных быстрорастущих или капиталоёмких компаний внутри этих рынков.

Для дальнейшего анализа эти компании стоит исключить из основной выборки, чтобы крупные единичные сделки не искажали оценку типичных объёмов финансирования, динамики рынка и сравнение сегментов.

#### 3.2.3. Анализ типов финансирования по объёму и популярности

Проанализируем типы финансирования по двум параметрам: суммарному объёму привлечённых средств и популярности среди компаний. Сравнение этих показателей позволит определить, какие типы финансирования встречаются чаще всего, какие привлекают наибольшие суммы, а также выявить различия между массовыми и крупными по объёму источниками финансирования.

Дополнительно рассмотрим суммарные возвраты по типам финансирования на основе отдельного датасета с данными о возвратах.

```python
# Посчитаем суммарный объём финансирования по каждому типу в млн $
funding_by_type = investments_clean[returns.columns].sum().sort_values(ascending=False) / 1_000_000
funding_by_type
```

```python
# Посчитаем, сколько раз каждый тип финансирования встречается в датасете
funding_count_by_type = (investments_clean[returns.columns] > 0).sum().sort_values(ascending=False)
funding_count_by_type
```

```python
# Посмотрим на данные в табличном виде
funding_type_summary = pd.DataFrame({
    'Объём финансирования, млн $': funding_by_type,
    'Кол-во компаний': funding_count_by_type,
    'Средний объём на компанию, млн $': funding_by_type / funding_count_by_type 
})

funding_type_summary = funding_type_summary.sort_values('Объём финансирования, млн $', ascending=False)
funding_type_summary
```

Второе место по популярности и объёму занимает `seed`, однако его средний объём на компанию значительно ниже - 0.71 млн долларов против 6.86 млн долларов у `venture`.

Некоторые типы финансирования используются относительно часто, но характеризуются небольшими средними объёмами. Например, `seed` и `angel` встречаются у большого количества компаний, однако средний объём на компанию по ним составляет 0.71 млн и 0.84 млн долларов соответственно. Это указывает на то, что такие инструменты чаще связаны с ранними стадиями развития компаний.

Обратная ситуация наблюдается у `private_equity`, этот тип финансирования встречается значительно реже, чем `venture`, `seed`, `debt_financing` и `angel`, но имеет высокий средний объём на компанию - 7.64 млн долларов. Ещё более высокий средний объём наблюдается у `post_ipo_equity` и `post_ipo_debt` - 11.87 млн и 10.62 млн долларов соответственно, но эти типы встречаются редко.

Наименее распространёнными типами финансирования являются `secondary_market`, `post_ipo_debt`, `post_ipo_equity` и `product_crowdfunding`. При этом `equity_crowdfunding` и `product_crowdfunding` характеризуются небольшими суммарными и средними объёмами финансирования.

Напротив, `private_equity`, `post_ipo_equity` и `post_ipo_debt` используются значительно реже, но отличаются высокими средними объёмами финансирования. Такие инструменты скорее связаны с более зрелыми компаниями и крупными сделками.

Результаты показывают, что разные типы финансирования выполняют разные функции. Для охвата большого количества компаний на ранних стадиях подходят `seed` и `angel`, но для получения более крупных возвратов и работы со зрелыми компаниями более перспективными выглядят `venture`, `private_equity` и `debt_financing`.

### 3.3. Анализ динамики <a id='section_3.3'></a>

#### 3.3.1. Динамика предоставления финансирования по годам

Проанализируем динамику финансирования стартапов по годам. Для этого рассчитаем средний объём одного раунда финансирования для каждой компании и оценим, как менялся типичный размер раунда во времени.

Дополнительно рассмотрим динамику общего количества раундов финансирования по годам. Это позволит оценить не только размер привлекаемых средств, но и общую активность инвестиционного рынка в разные периоды.

```python
# Рассчитаем средний объём одного раунда финансирования для каждой компании
investments_clean['avg_funding_per_round'] = investments_clean['funding_total_usd'] / investments_clean ['funding_rounds']
```

```python
# Изучим статистические показатели столбца avg_funding_per_round
investments_clean['avg_funding_per_round'].describe()
```

```python
# Посмотрим на динамику среднего объёма одного раунда и общего количества раундов
yearly_dynamics = (
    investments_clean
    .groupby('mid_funding_year')
    .agg(
        avg_funding_per_round=('avg_funding_per_round', 'mean'),
        median_funding_per_round=('avg_funding_per_round', 'median'),
        total_funding_rounds=('funding_rounds', 'sum'))
)

yearly_dynamics
```

```python
# Построим график динамики типичного размера одного раунда финансирования
ax = sns.lineplot(
    data=yearly_dynamics,
    x=yearly_dynamics.index,
    y=yearly_dynamics['median_funding_per_round'] / 1_000_000,
    marker='o',
    color='#66c2a5'
)

# Добавим заголовок и подписи осей
ax.set_title('Динамика типичного размера одного раунда финансирования')
ax.set_xlabel('Год')
ax.set_ylabel('Медианный объём одного раунда, млн $')

# Отобразим график
plt.show()
```

```python
# Построим график динамики количества раундов финансирования
ax = sns.lineplot(
    data=yearly_dynamics,
    x=yearly_dynamics.index,
    y='total_funding_rounds',
    marker='o',
    color='#6b9ac4'
)

# Добавим заголовок и подписи осей
ax.set_title('Динамика количества раундов финансирования')
ax.set_xlabel('Год')
ax.set_ylabel('Количество раундов')

# Отобразим график
plt.show()
```

```python
# Найдём год с максимальным типичным размером одного раунда
max_median_round_year = yearly_dynamics['median_funding_per_round'].idxmax()
max_median_round_value = yearly_dynamics['median_funding_per_round'].max() / 1_000_000

display(f'Максимальный медианный объём одного раунда был в {max_median_round_year} году')
display(f'Медианный объём одного раунда: {round(max_median_round_value, 2)} млн долларов')
```

Максимальный типичный размер средств, привлечённых в рамках одного раунда финансирования, наблюдался в 2005 году. В этот период медианный объём одного раунда достиг примерно 4.5 млн долларов.

После 2005–2006 годов медианный размер раунда постепенно снижался и к 2013–2014 годам находился уже на уровне около 0.56 млн долларов.

При этом количество раундов финансирования, наоборот, активно росло, с 948 раундов в 2005 году до 12 877 раундов в 2013 году. Это говорит о расширении инвестиционного рынка, финансирование стало получать большее количество компаний, но типичный объём одного раунда стал меньше. Можно предположить что в 2005 году произошло насыщение рынка крупных стартапов, и финансироваться стали даже самые мелкие. Это объясняет рост рандов и снижение финансирования за раунд.

В 2014 году количество раундов снизилось до 7 116, а медианный размер одного раунда остался почти на уровне 2013 года (около 0.56 млн долларов). Однако снижение количества раундов в 2014 году стоит интерпретировать осторожно, поскольку декабрь представлен заметно меньшим количеством записей, и у нас нет полной уверенности в полноте данных за этот месяц.

После 2006 года рынок стал более массовым и менее концентрированным по размеру сделок, инвестиции распределялись между большим числом компаний, но средний чек типичного раунда снижался. Это может означать рост конкуренции за перспективные стартапы и смещение рынка в сторону более ранних или менее капиталоёмких сделок.

#### 3.3.2. Динамика размера общего финансирования по массовым сегментам рынка для растущих в 2014 году сегментов

Проанализируем динамику общего объёма финансирования по рыночным сегментам, для этого построим сводную таблицу по годам и сегментам, затем отберём только те сегменты, где суммарный объём финансирования в 2014 году вырос по сравнению с 2013 годом.

Далее ограничим анализ массовыми сегментами, так как по ним достаточно наблюдений для более устойчивой интерпретации динамики.

```python
# Создаём сводную таблицу объёма финансирования по годам и сегментам
market_year_pivot = pd.pivot_table(investments_clean,
                                   index='mid_funding_year',
                                   columns='market',
                                   values='funding_total_usd',
                                   aggfunc='sum'
)

market_year_pivot
```

```python
# Определим сегменты, которые выросли в 2014 году по сравнению с 2013
values_change = market_year_pivot.loc[2014] - market_year_pivot.loc[2013]
growing_markets_2014 = values_change.loc[values_change > 0].sort_values(ascending=False)

# Оставим только массовые сегменты
growing_markets_2014 = (growing_markets_2014.drop(['niche', 'mid', 'unknown']))

growing_markets_2014
```

```python
# Построим график динамики финансирования по сегментам
ax = (market_year_pivot[growing_markets_2014.index] / 1_000_000).plot(
    figsize=(12, 7),
    marker='o',
    markersize=4,
    linewidth=1,
    color=sns.color_palette('deep')
)

# Добавим заголовок и подписи осей
ax.set_title('Динамика финансирования растущих массовых сегментов')
ax.set_xlabel('Год')
ax.set_ylabel('Суммарный объём финансирования, млн $')
ax.legend(title='Сегмент')

# Отобразим график
plt.show()
```

Среди растущих массовых сегментов самый крупный объём финансирования показывает `manufacturing`, в 2014 году он достиг 416.3 млн долларов. Однако динамика сегмента волатильна, заметны резкие скачки и спады, поэтому его рост нельзя считать полностью устойчивым.

Наиболее уверенный рост наблюдается у `medical`, `internet`, `real estate` и `saas`. Эти сегменты увеличивали объём финансирования к 2014 году более плавно и последовательно.

Самый быстрый рост в 2014 году относительно 2013 года показал `medical`, также заметный рост наблюдается у `technology`, `internet`, `apps`, `startups`, `real estate` и `manufacturing`.

Наиболее интересными выглядят сегменты, где рост сочетается с относительной стабильностью: `medical`, `internet`, `real estate` и `saas`. Они показывают не только рост в 2014 году относительно 2013 года, но и более понятную долгосрочную положительную динамику.

#### 3.3.3. Годовая динамика доли возвращённых средств по типам финансирования

Проанализируем годовую динамику доли возвращённых средств по основным типам финансирования. Для этого сравним объём возвратов из дополнительного датасета с суммарным объёмом предоставленного финансирования за каждый год.

Такой показатель позволит оценить, какая часть вложенных или выданных средств возвращалась инвесторам и финансистам, а также сравнить устойчивость динамики по разным типам финансирования.

```python
# Посчитаем суммарный объём предоставленного финансирования по годам и типам в млн $
funding_sum = (
    investments_clean
    .groupby('mid_funding_year')[returns.columns]
    .sum() / 1_000_000
)

funding_sum 
```

```python
# Рассчитаем долю возвращённых средств от предоставленных
return_share = returns / (funding_sum + 1e-60)
```

Так как показатель отражает долю возвращённых средств от предоставленного финансирования, значения выше 1 считаем неадекватно большими для интерпретации в рамках одного года, поэтому заменим их на пропуски.

```python
# Заменяем неадекватно большие значения на пропуски
return_share = return_share.mask(return_share > 1)

return_share
```

```python
# Оставим только выбранные типы финансирования
main_return_share = return_share[['venture', 'debt_financing', 'private_equity', 'seed', 'angel']]
```

```python
# Построим график годовой динамики доли возвращённых средств
ax = (main_return_share * 100).plot(
    marker='o',
    markersize=4,
    linewidth=1,
)

# Добавим заголовок и подписи осей
ax.set_title('Динамика доли возвращённых средств по типам финансирования')
ax.set_xlabel('Год')
ax.set_ylabel('Доля возвращённых средств, %')
ax.legend(title='Тип финансирования', bbox_to_anchor=(1.02, 1))

# Отобразим график
plt.show()
```

Среди рассматриваемых типов финансирования наиболее заметная положительная динамика наблюдается у `venture`. После снижения в 2006–2009 годах доля возвращённых средств постепенно растёт и к 2014 году достигает одного из максимальных уровней за период.

У `debt_financing` динамика менее стабильная: показатель заметно колеблется, а в 2014 году значение было заменено на пропуск как аномально высокое. Поэтому устойчивый рост по этому типу финансирования не прослеживается.

`private_equity` показывает относительно высокие значения в 2009–2013 годах, но динамика также нестабильна, а значение за 2014 год было исключено как выброс. `seed` и `angel` не показывают устойчивого роста: их доля возвратов меняется волнообразно и в последние годы скорее снижается или остаётся нестабильной.

Таким образом, наиболее уверенная положительная динамика среди рассматриваемых типов финансирования наблюдается у `venture`. Рост или снижение доли возвращённых средств не всегда напрямую отражает эффективность конкретного типа финансирования. В датасете отсутствуют показатели инвестиционной доходности (ROI), а на динамику возвратов могут влиять временной лаг между инвестициями и возвратами и отдельные крупные сделки. Поэтому полученные результаты скорее отражают особенности распределения возвратов между типами финансирования.
